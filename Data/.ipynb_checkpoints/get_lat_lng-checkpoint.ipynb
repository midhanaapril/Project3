```python
df2 = df[df['street_address'].notna()]
df2
```

```python
addresses = df2['street_address'] + " " + df2['city'] + " " + df2['state']
address = addresses.tolist()
address

```

```python
API_KEY = 'AIzaSyB9Y5b2UyHiMSj0e3U8jJiWNZJxE5u5ndQ'
base_url = 'https://maps.googleapis.com/maps/api/geocode/json?'
lat = []
lng = []
confirmed = []
for x in address:
    query = base_url + x
    params = {
    'key': API_KEY,
    'address': x
    } 
    
    
    response = requests.get(base_url, params=params).json()
    
    geometry = response['results'][0]['geometry']
    lat.append(geometry['location']['lat'])
    lng.append(geometry['location']['lng'])
    confirmed.append(x)
    print(f"{x} found! Appending stats")
```
